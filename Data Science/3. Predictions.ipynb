### Modelling

```python
# Subset to only quantitative vars
num_vars = results[['Salary', 'CareerSatisfaction', 'HoursPerWeek', 'JobSatisfaction', 'StackOverflowSatisfaction']]


num_vars.head()

# We need to handle missing values otherwise sklearn will break
```

```python
# Remove all nans
all_rm = num_vars.dropna() 

all_rm.head()
```

```python
# Consider only numerical variables
X = all_rm[['CareerSatisfaction', 'HoursPerWeek', 'JobSatisfaction', 'StackOverflowSatisfaction']]
y = all_rm['Salary']

# Split data into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state=42)
# Instantiate
lm_model = LinearRegression(normalize=True) 
# Fit
lm_model.fit(X_train, y_train) 
```

```python
# Use lm_model to predict y_test response values
y_test_preds = lm_model.predict(X_test)
# obtain an r-squared value for how well the predicted values compare to the actual test values
r2_test = r2_score(y_test, y_test_preds)
r2_test

# This means 3% variability in response can be explained by our model
```
